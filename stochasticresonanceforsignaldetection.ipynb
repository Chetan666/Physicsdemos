```python
import numpy as np
import matplotlib.pyplot as plt

def subthreshold_signal(t, frequency=1, amplitude=0.5):
    return amplitude * np.sin(2 * np.pi * frequency * t)

def add_noise(signal, noise_level=0.2):
    noise = noise_level * np.random.normal(size=len(signal))
    return signal + noise

def stochastic_resonance(signal, noise_level, threshold=0.5):
    noisy_signal = add_noise(signal, noise_level)
    output_signal = np.where(noisy_signal > threshold, 1, 0)
    return noisy_signal, output_signal

# Parameters
t = np.linspace(0, 10, 1000)  # Time vector
frequency = 1  # Frequency of the subthreshold signal
amplitude = 0.5  # Amplitude of the subthreshold signal
noise_level = 1  # Level of noise

# Generate subthreshold signal
input_signal = subthreshold_signal(t, frequency, amplitude)

# Apply stochastic resonance
noisy_signal, output_signal_with_resonance = stochastic_resonance(input_signal, noise_level)

# Apply threshold without stochastic resonance
output_signal_without_resonance = np.where(input_signal > 0.5, 1, 0)

# Plotting
plt.figure(figsize=(10, 8))

plt.subplot(4, 1, 1)
plt.plot(t, input_signal, label='Subthreshold Signal')
plt.title('Subthreshold Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(t, output_signal_without_resonance, label='Output Signal (without Stochastic Resonance)', color='orange')
plt.title('Output Signal (without Stochastic Resonance)')
plt.xlabel('Time')
plt.ylabel('Binary Output')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(t, noisy_signal, label='Noisy Signal (with Stochastic Resonance)')
plt.title('Noisy Signal (with Stochastic Resonance)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(t, output_signal_with_resonance, label='Output Signal (with Stochastic Resonance)')
plt.title('Output Signal (with Stochastic Resonance)')
plt.xlabel('Time')
plt.ylabel('Binary Output')
plt.legend()



plt.tight_layout()
plt.show()
```
